Given an integer¬†`n`, return¬†*`true`¬†if it is a power of two. Otherwise, return¬†`false`*.

An integer¬†`n`¬†is a power of two, if there exists an integer¬†`x`¬†such that¬†`n == 2x`.

**Example 1:**
Input: n = 1 

Output: true

**Example 2:**
Input: n = 16 

Output: true

**Example 3:**
Input: n = 3 

Output: false

</aside>

```python
def power_2(n):
    if n==1:
        return True
    if (n/2)==1:
        return True
    elif (n/2)<2:
        return False
    return power_2(n/2)
power_2(24)
```

**Example 1:**

Input: N = 5 

Output: 120

**Example 2:**

Input: N = 4

</aside>

```python
def factorial(n):
    fact=1
    fact*=n
    if n==1:
        return fact
    return fact*factorial(n-1)
factorial(5)
```

Given a number N and a power P, the task is to find the exponent of this number raised to the given power, i.e. N^P.

**Example 1 :** 

Input: N = 5, P = 2

Output: 25

**Example 2 :**
Input: N = 2, P = 5

Output: 32

</aside>

```python
def power(n,p):
    result=1
    result*=n
    if p==1:
        return result
    return result*power(n,p-1)
```

Given first term (a), common difference (d) and a integer N of the Arithmetic Progression series, the task is to find Nth term of the series.

**Example 1:**

Input : a = 2 d = 1 N = 5
Output : 6
The 5th term of the series is : 6

**Example 2:**

Input : a = 5 d = 2 N = 10
Output : 23
The 10th term of the series is : 23

```python
def nterm(a,d,N):
    term=0
    term+=d
    if N==2:
        return a+term
    return d+nterm(a,d,N-1)
```

üí° **Question 7**

Given a string S, the task is to write a program to print all permutations of a given string.

**Example 1:**

***Input:***

*S = ‚ÄúABC‚Äù*

***Output:***

*‚ÄúABC‚Äù, ‚ÄúACB‚Äù, ‚ÄúBAC‚Äù, ‚ÄúBCA‚Äù, ‚ÄúCBA‚Äù, ‚ÄúCAB‚Äù*

**Example 2:**

***Input:***

*S = ‚ÄúXY‚Äù*

***Output:***

*‚ÄúXY‚Äù, ‚ÄúYX‚Äù*

</aside>

```python
def permutations(S):
    # Base case: If the string has only one character, return it as a single permutation
    if len(S) == 1:
        return [S]
    
    # Recursive case: Generate permutations by fixing one character and permuting the rest
    result = []
    for i in range(len(S)):
        # Fix the i-th character as the first character
        fixed_char = S[i]
        
        # Generate permutations for the remaining characters
        remaining_chars = S[:i] + S[i+1:]
        sub_permutations = permutations(remaining_chars)
        
        # Append the fixed character to each permutation of the remaining characters
        for sub_permutation in sub_permutations:
            result.append(fixed_char + sub_permutation)
    
    return result


permutations("ABC")

```

Given an array, find a product of all array elements.

**Example 1:**

Input  : arr[] = {1, 2, 3, 4, 5}
Output : 120
**Example 2:**

Input  : arr[] = {1, 6, 3}
Output : 18

</aside>

```python
def prod_array(arra):
    result=1
    b=(len(arra)-1)
    result*=arra[b]
    if len(arra)==1:
        return result
    arra.pop()
    return result*prod_array(arra)
```
